**Задание 1.** Вычислите значение выражения $\displaystyle\sqrt[5]{\frac{\ln2}{{(\sqrt{e}-1)}^2{(\sin3+4)}^3}+\sqrt[8]{e^3}}$ с точностью 40 значащих десятичных цифр.

```python
from sympy import *

a = (log(2)) / (((E**(S(1)/2) - 1)**(2)) * ((sin(3) + 4)**(3))) + E**(S(3)/8)
b = a**(S(1)/5)

N(S(b), 40)
```

**Задание 2.** Раскройте скобки в выражении $\left(u + z\right) \left(x - 2 y\right)^{2} \left(2 x - y\right)^{2}$.

**Задание 4.** Найдите все (включая комплексные) корни уравнения $35 x^{4} + 31 x^{3} + 118 x^{2} - 262 x - 84 = 0$.

**Задание 5.** Постройте график функции, заданной неявно: $\displaystyle - x^{3} - 2 x^{2} - x + y^{2}=0$. Используйте одинаковый масштаб по координатным осям.

```python
from math import * #5
var('y')
q = (-(x**3)) - 2*(x**2) - x + (y**2)
plot_implicit(q, (x, -2, 2), (y, -2, 2), aspect_ratio = (5,5))
```

**Задание 6.** Постройте график функции, заданной параметрически: $\displaystyle x(t)=6 \sin{\left(4 t \right)}$, $\displaystyle y(t)=4 \sin{\left(3 t \right)}$. Используйте минимальный по длине промежуток изменения параметра и одинаковый масштаб по координатным осям.

```python
from sympy import * #6
var('t')
plotting.plot_parametric((6*sin(4*t), 4*sin(3*t)), (t, 0, 2 * pi), aspect_ratio = (1,1))
```
